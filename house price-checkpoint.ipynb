```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('home_dataset.csv')
```

```python
house_sizes = data['HouseSize'].values
house_prices = data['HousePrice'].values
```

```python
plt.scatter(house_prices, house_sizes, marker='o', color='blue')
plt.title('House Price vs House Size')
plt.xlabel('House Size(sq.ft)')
plt.ylabel('House Price($)')
plt.show()
            
```

```python
x_train, x_test, y_train, y_test = train_test_split(house_prices, house_sizes, test_size=0.2, random_state=42)
```

```python
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)

model = LinearRegression()
model.fit(x_train, y_train)

```

```python
# Predict prices for the test set
predictions = model.predict(x_test)

# Visualize the predictions
```

```python
plt.scatter(x_test, y_test, marker='o', color='blue', label='Actual Prices')
plt.plot(x_test, predictions, color='red', linewidth=2, label='Predicted Prices')
plt.title('Dumbo Property Price Prediction with Linear Regression')
plt.xlabel('House Size (sq.ft)')
plt.ylabel('House Price (millions $)')
plt.legend()
plt.show()
```

```python
# --- Predict price for a custom house size ---

house_size = float(input("Enter the size of the house (in sq ft): "))

# Reshape input for model
house_size_input = np.array([[house_size]])

# Predict
predicted_price = model.predict(house_size_input)

print(f"Predicted Price for a {house_size} sq ft house: ${predicted_price[0]:.2f} million")


```
